#  QUESTION 5: Applied Logistic Regression

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
xtrain = pd.read_csv("XTrain.csv",header=0)
ytrain =pd.read_csv("yTrain.csv",header=0)
xtest = pd.read_csv("XTest.csv",header=0)
ytest = pd.read_csv("ytest.csv",header=0)
vocab = pd.read_csv("vocabulary.csv",header=None)
```

## Q1:Misclassification test rate and Confusion Matrix

```python
#Part a
#Misclassification Rate
#Confusion Mattrix
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as skm
logistic = LogisticRegression(C=1e24)
model = logistic.fit(xtrain,ytrain)
y_pred_train = model.predict(xtrain)
y_pred_test = model.predict(xtest)
test_score = skm.accuracy_score(ytest,y_pred_test)
test_error =1-test_score
print("Misclassification rate on test data is",test_error)
conf = skm.confusion_matrix(ytest,y_pred_test)
print("Confusion Mattrix:")
print(conf)
```

## Q2:"Precision-Recall Curve"

```python
#Part b
#Recall precision and ROC curve

scores = logistic.predict_proba(xtest)[:,1]
precision, recall, _ = skm.precision_recall_curve(ytest, scores)
average_precision = skm.average_precision_score(ytest, scores)

plt.plot(recall, precision, label='area = %0.2f' % average_precision, color="green")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision Recall Curve')
plt.legend(loc="lower right")
fpr,tpr,_=skm.roc_curve(ytest,scores)
plt.show()
plt.plot(fpr,tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive rate')
plt.ylabel('False Negative Rate')
plt.title('ROC Curve')
plt.show()
```

## Q3:"Word Cloud"

```python
#conda install -c conda-forge wordcloud
from wordcloud import WordCloud, STOPWORDS
d={}
absd={}
coff = pd.DataFrame(logistic.coef_.T)
coff.columns = ['coff']
words = pd.DataFrame(vocab)
words.columns=['words']
result = pd.concat([coff, words], axis=1, join='inner')
result.set_index('words')
lst = result.to_dict("records")
for i in lst:
    d[i['words']]=i['coff']
    absd[i['words']]=abs(i['coff'])
```

```python
def grey_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    v = d[word]
    if v>0:
        (h,s,l) =(90,100,49)
        return "hsl({}, {}%, {}%)".format(h, s, l)
    else:
        (h,s,l) = (0, 100, 50)
        return "hsl({}, {}%, {}%)".format(h, s, l)

wordcloud = WordCloud(width=900,height=500,normalize_plurals=False,color_func=grey_color_func).generate_from_frequencies(absd)
#wordcloud.recolor(color_func = grey_color_func)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
print("We see that the green words are larger in size that means they have higher positive coffecient and they are more in number which means there are more larger positive coffecient words in our dataset.")
```

## Q4:Lambda v/s Cross Validation plot (L1 REGULARIZATION)

```python
import sklearn.linear_model as slm
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split

classifier = slm.LogisticRegressionCV(penalty='l1',solver="liblinear")
clf = classifier.fit(xtrain,ytrain)
# print('Coefficient of each feature:', classifier.coef_)
# print("Scores",classifier.scores_)
# print('Training accuracy:', classifier.score(xtrain, ytrain))
# print('Test accuracy:', classifier.score(xtest, ytest))
# print("missclassification:",1-classifier.score(xtest, ytest))
# print("Hyperparameters",classifier.Cs_)
# print("c_",classifier.C_)
```

```python
from sklearn.model_selection import GridSearchCV
import sklearn.linear_model as slm
from sklearn.model_selection import RandomizedSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10,100,1000] }
model =slm.LogisticRegression(penalty='l1',solver="liblinear")
clff = GridSearchCV(model, param_grid)
```

```python
y = np.array(ytrain).reshape(579,)
clff.fit(xtrain.values,y)
print(clff.best_score_)
print(clff.best_params_)
print(clff.best_estimator_.C)
```

```python
print("We find the most optimal value of C:1000 with max accuracy of 95.6%, hence lambda(optimal) is : 0.001")
print("Below is a  table of all the possible values of C iterated over")
print("Below is also a lambda vs cross validation error  graph plotted")
```

```python
data=clff.cv_results_
df = pd.DataFrame()
df= df.from_dict(data,orient='columns')
cross_val_error = 1-df["mean_test_score"]
lambdaa=1/np.array(df["param_C"])
plt.plot(lambdaa,cross_val_error)
plt.title("Lambda V/S Cross Val Error")
plt.xlabel('Lambda')
plt.ylabel('Cross Val Error')
plt.show()
df
```

## Q5:Missclassifcation rate and Confusion mattrix (L1 REGULARIZATION)

```python
classifier = slm.LogisticRegression(C=1000, penalty='l1',solver="liblinear")
model = classifier.fit(xtrain,ytrain)
y_pred_train = model.predict(xtrain)
y_pred_test = model.predict(xtest)
test_score = skm.accuracy_score(ytest,y_pred_test)
test_error =1-test_score
print("Misclassification rate on test data is",test_error)
conf = skm.confusion_matrix(ytest,y_pred_test)
print("Confusion Mattrix:")
print(conf)
```

```python
print("Yes on choosing the optimal value of lamda we get a lowered test error from 0.048 to 0.041")
```

## Q6:Precision and Recall curve(L1 REGULARIZATION)

```python
scores = classifier.predict_proba(xtest)[:,1]
precision, recall, _ = skm.precision_recall_curve(ytest, scores)
average_precision = skm.average_precision_score(ytest, scores)

plt.plot(recall, precision, label='area = %0.2f' % average_precision, color="green")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision Recall Curve')
plt.legend(loc="lower right")
fpr,tpr,_=skm.roc_curve(ytest,scores)
plt.show()
plt.plot(fpr,tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive rate')
plt.ylabel('False Negative Rate')
plt.title('ROC Curve')
plt.show()
```

## Q7:"Word Cloud"(L1 regularization)

```python
#conda install -c conda-forge wordcloud
from wordcloud import WordCloud, STOPWORDS
d2={}
absd2={}
coff = pd.DataFrame(classifier.coef_.T)
coff.columns = ['coff']
words = pd.DataFrame(vocab)
words.columns=['words']
result = pd.concat([coff, words], axis=1, join='inner')
result.set_index('words')
lst = result.to_dict("records")
for i in lst:
    d2[i['words']]=i['coff']
    absd2[i['words']]=abs(i['coff'])
    
def grey_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    v = d2[word]
    if v>0:
        (h,s,l) =(90,100,49)
        return "hsl({}, {}%, {}%)".format(h, s, l)
    else:
        (h,s,l) = (0, 100, 50)
        return "hsl({}, {}%, {}%)".format(h, s, l)

wordcloud = WordCloud(width=900,height=500,normalize_plurals=False,color_func=grey_color_func).generate_from_frequencies(absd2)
#wordcloud.recolor(color_func = grey_color_func)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
